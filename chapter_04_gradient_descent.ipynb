# Chapter 4: Introduction to Neural Learning
### Gradient Descent

This chapter covers the **compare** and **learn** steps. After a prediction is made, the network measures how wrong it was and then adjusts its weights to reduce that error.

## 1. Measuring Error: Mean Squared Error

The network needs a number that tells it how much it missed. Mean Squared Error (MSE) is the most common choice. Squaring the raw error makes it positive and amplifies large mistakes while suppressing small ones.

````python
knob_weight = 0.5
input_val   = 0.5
goal_pred   = 0.8

pred  = input_val * knob_weight
error = (pred - goal_pred) ** 2

print(f"Prediction : {pred}")
print(f"Goal       : {goal_pred}")
print(f"Raw error  : {pred - goal_pred:.4f}")
print(f"MSE        : {error:.4f}")
````

## 2. Hot and Cold Learning

The simplest learning strategy: try moving the weight up and down by a fixed amount, keep whichever direction reduces the error. This is slow and imprecise but illustrates the core idea.

````python
weight          = 0.5
input_val       = 0.5
goal_prediction = 0.8
step_amount     = 0.001

for iteration in range(1101):
    prediction = input_val * weight
    error      = (prediction - goal_prediction) ** 2

    up_pred   = input_val * (weight + step_amount)
    up_error  = (goal_prediction - up_pred) ** 2

    down_pred  = input_val * (weight - step_amount)
    down_error = (goal_prediction - down_pred) ** 2

    if down_error < up_error:
        weight -= step_amount
    else:
        weight += step_amount

    if iteration % 200 == 0:
        print(f"Iteration {iteration:>4}: error={error:.8f}  prediction={prediction:.6f}")

print(f"\nFinal weight: {weight:.4f}")
````

## 3. Gradient Descent

A far better method: compute the direction and amount of the weight update in a single step using the derivative. The key formula is:

```
weight_delta = (prediction - goal) * input
weight      -= alpha * weight_delta
```

Multiplying by `input` provides three effects: **stopping** (no update when input=0), **negative reversal** (flips sign when input is negative), and **scaling** (larger inputs cause larger updates).

````python
weight    = 0.5
goal_pred = 0.8
input_val = 0.5

print(f"{'Iter':<6} {'Prediction':>12} {'Error':>14}")
print("-" * 36)

for iteration in range(20):
    pred                 = input_val * weight
    error                = (pred - goal_pred) ** 2
    direction_and_amount = (pred - goal_pred) * input_val
    weight              -= direction_and_amount
    print(f"{iteration:<6} {pred:>12.6f} {error:>14.8f}")
````

## 4. The Role of Alpha (Learning Rate)

Alpha controls how aggressively weights are updated. Too large and the network overshoots and diverges. Too small and it learns slowly.

````python
import numpy as np

def train(alpha, iterations=30):
    weight    = 0.0
    input_val = 0.5
    goal      = 0.8
    errors    = []
    for _ in range(iterations):
        pred   = input_val * weight
        error  = (pred - goal) ** 2
        delta  = pred - goal
        weight -= alpha * delta * input_val
        errors.append(round(error, 6))
    return errors

for alpha in [0.01, 0.1, 1.0, 10.0]:
    errors = train(alpha)
    status = "diverged" if errors[-1] > errors[0] else f"final error={errors[-1]}"
    print(f"alpha={alpha:<6} -> {status}")
````

## 5. Full Gradient Descent with Named Variables

The canonical single-weight implementation using `delta` and `weight_delta` as explicit intermediate variables, as the book introduces them.

````python
def neural_network(input_val, weight):
    return input_val * weight

number_of_toes     = [8.5]
win_or_lose_binary = [1]

weight    = 0.1
alpha     = 0.01
input_val = number_of_toes[0]
goal_pred = win_or_lose_binary[0]

print(f"{'Iter':<6} {'Weight':>10} {'Pred':>10} {'Error':>12} {'Delta':>10}")
print("-" * 55)

for iteration in range(15):
    pred         = neural_network(input_val, weight)
    error        = (pred - goal_pred) ** 2
    delta        = pred - goal_pred
    weight_delta = input_val * delta
    weight      -= alpha * weight_delta
    print(f"{iteration:<6} {weight:>10.6f} {pred:>10.6f} {error:>12.8f} {delta:>10.6f}")
````
