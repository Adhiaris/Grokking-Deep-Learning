# Chapter 3: Introduction to Neural Prediction
### Forward Propagation

This chapter covers the **predict** step of the Predict → Compare → Learn paradigm. Forward propagation is the process of sending input data through a neural network to produce a prediction.

## 1. Simplest Neural Network: Single Input, Single Output

A neural network with one input and one weight multiplies the input by the weight to produce a prediction. The weight acts like a volume knob — scaling the input up or down.

```python
weight = 0.1

def neural_network(input_val, weight):
    return input_val * weight

number_of_toes = [8.5, 9.5, 10.0, 9.0]

print(f"{'Input (avg toes)':<20} {'Prediction (win prob)'}")
print("-" * 42)
for toes in number_of_toes:
    pred = neural_network(toes, weight)
    print(f"{toes:<20} {pred:.4f}")
```

## 2. Multiple Inputs, Single Output (Weighted Sum / Dot Product)

When a network has multiple inputs, each input is multiplied by its own weight and the results are summed. This is called a **weighted sum** or **dot product**. It measures similarity between the input vector and the weight vector.

```python
import numpy as np

weights = np.array([0.1, 0.2, 0.0])

def neural_network(input_vec, weights):
    return input_vec.dot(weights)

toes  = np.array([8.5, 9.5, 9.9, 9.0])
wlrec = np.array([0.65, 0.8, 0.8, 0.9])
nfans = np.array([1.2, 1.3, 0.5, 1.0])

input_vec = np.array([toes[0], wlrec[0], nfans[0]])
pred = neural_network(input_vec, weights)

print("Input vector:", input_vec)
print("Weights:     ", weights)
print()
print("Weighted contributions:")
labels = ["toes", "wlrec", "nfans"]
for label, inp, w in zip(labels, input_vec, weights):
    print(f"  {label}: {inp} * {w} = {inp*w:.4f}")
print(f"\nFinal prediction (sum): {pred:.4f}")
```

## 3. Dot Product Properties

The dot product is a measure of similarity between two vectors. Vectors that overlap in the same positions produce high scores. Negative weights act as a NOT operator.

```python
import numpy as np

a = np.array([0, 1, 0, 1], dtype=float)
b = np.array([1, 0, 1, 0], dtype=float)
c = np.array([0, 1, 1, 0], dtype=float)
d = np.array([0.5, 0, 0.5, 0], dtype=float)
e = np.array([0, 1, -1, 0], dtype=float)

pairs = [
    ("a", "b", a, b),
    ("b", "c", b, c),
    ("c", "c", c, c),
    ("c", "e", c, e),
    ("d", "d", d, d),
]

print(f"{'Pair':<12} {'Score'}")
print("-" * 20)
for n1, n2, v1, v2 in pairs:
    score = v1.dot(v2)
    print(f"dot({n1},{n2})     {score}")
```

## 4. Single Input, Multiple Outputs

One input can produce multiple predictions via elementwise multiplication with a weight vector. Each output is independent.

```python
import numpy as np

weights = np.array([0.3, 0.2, 0.9])

def neural_network(input_val, weights):
    return input_val * weights

wlrec = [0.65, 0.8, 0.8, 0.9]
input_val = wlrec[0]
pred = neural_network(input_val, weights)

print(f"Input (win/loss record): {input_val}")
print(f"Weights: {weights}")
print()
print(f"Prediction (hurt prob): {pred[0]:.4f}")
print(f"Prediction (win  prob): {pred[1]:.4f}")
print(f"Prediction (sad  prob): {pred[2]:.4f}")
```

## 5. Multiple Inputs, Multiple Outputs (Vector-Matrix Multiplication)

The most general single-layer case: each output is a dot product of the input vector with one row of the weight matrix.

```python
import numpy as np

weights = np.array([
    [0.1,  0.1, -0.3],
    [0.1,  0.2,  0.0],
    [0.0,  1.3,  0.1],
]).T

def neural_network(input_vec, weights):
    return input_vec.dot(weights)

toes  = np.array([8.5, 9.5, 9.9, 9.0])
wlrec = np.array([0.65, 0.8, 0.8, 0.9])
nfans = np.array([1.2, 1.3, 0.5, 1.0])

input_vec = np.array([toes[0], wlrec[0], nfans[0]])
pred = neural_network(input_vec, weights)

print("Input:", input_vec)
print(f"Prediction (hurt): {pred[0]:.4f}")
print(f"Prediction (win) : {pred[1]:.4f}")
print(f"Prediction (sad) : {pred[2]:.4f}")
```

## 6. Predicting on Predictions (Stacking Layers)

Output from one network can feed into another — this creates a hidden layer. The first layer learns intermediate representations; the second layer uses them to produce the final prediction.

```python
import numpy as np

ih_wgt = np.array([
    [ 0.1,  0.2, -0.1],
    [-0.1,  0.1,  0.9],
    [ 0.1,  0.4,  0.1],
]).T

hp_wgt = np.array([
    [0.3,  1.1, -0.3],
    [0.1,  0.2,  0.0],
    [0.0,  1.3,  0.1],
]).T

weights = [ih_wgt, hp_wgt]

def neural_network(input_vec, weights):
    hid  = input_vec.dot(weights[0])
    pred = hid.dot(weights[1])
    return pred

toes  = np.array([8.5, 9.5, 9.9, 9.0])
wlrec = np.array([0.65, 0.8, 0.8, 0.9])
nfans = np.array([1.2, 1.3, 0.5, 1.0])

input_vec = np.array([toes[0], wlrec[0], nfans[0]])
pred = neural_network(input_vec, weights)

print("Two-layer network predictions:")
print(f"  hurt: {pred[0]:.4f}")
print(f"  win : {pred[1]:.4f}")
print(f"  sad : {pred[2]:.4f}")
```

## 7. NumPy Shape Rules for Matrix Multiplication

The golden rule: when dotting two matrices, the inner dimensions must match. `(a, b).dot(b, c)` outputs `(a, c)`.

```python
import numpy as np

shapes = [
    ((1, 4), (4, 3)),
    ((2, 4), (4, 3)),
    ((2, 1), (1, 3)),
]

print(f"{'Left shape':<15} {'Right shape':<15} {'Output shape'}")
print("-" * 45)
for s1, s2 in shapes:
    a = np.zeros(s1)
    b = np.zeros(s2)
    c = a.dot(b)
    print(f"{str(s1):<15} {str(s2):<15} {c.shape}")
```
