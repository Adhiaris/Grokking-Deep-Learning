# Chapter 2: Fundamental Concepts
### How Do Machines Learn?

This chapter introduces the taxonomy of machine learning. Before building neural networks, you need to understand where they fit in the broader landscape of AI and ML.

## 1. The AI / ML / DL Hierarchy

Deep learning is a subset of machine learning, which is a subset of artificial intelligence. Every deep learning algorithm is a machine learning algorithm, but not every machine learning algorithm is a deep learning algorithm.

```python
hierarchy = [
    ("Artificial Intelligence", "Systems that mimic human intelligence"),
    ("  Machine Learning", "Systems that learn from data"),
    ("    Deep Learning", "ML using layered neural networks"),
]

for name, desc in hierarchy:
    print(f"{name:<30} {desc}")
```

## 2. Supervised vs. Unsupervised Learning

All machine learning splits into two flavors based on whether a 'right answer' exists during training.

- **Supervised**: Learn to transform input dataset A into output dataset B. A 'true' label exists for every training example.
- **Unsupervised**: Find structure in a single dataset without labels. The result is typically a grouping (clustering) of the data.

```python
import numpy as np

supervised_examples = [
    ("Image pixels", "Cat or not cat"),
    ("Movie ratings", "Movie recommendations"),
    ("Audio file", "Text transcript"),
    ("Weather sensors", "Rain probability"),
    ("Monday stock price", "Tuesday stock price"),
]

print("Supervised Learning Examples")
print(f"{'Input (what you know)':<30} {'Output (what you want to know)'}")
print("-" * 65)
for inp, out in supervised_examples:
    print(f"{inp:<30} {out}")
```

```python
words = ["puppies", "pizza", "kittens", "hot dog", "burger"]

cluster_map = {
    "puppies": 1,
    "pizza": 2,
    "kittens": 1,
    "hot dog": 2,
    "burger": 2,
}

print("Unsupervised Learning: Clustering Example")
print(f"{'Word':<12} {'Cluster'}")
print("-" * 25)
for word in words:
    label = cluster_map[word]
    meaning = "cute" if label == 1 else "delicious"
    print(f"{word:<12} {label}  ({meaning})")

print("\nThe algorithm found structure without being told what the clusters mean.")
```

## 3. Parametric vs. Nonparametric Learning

A second axis divides machine learning into two more groups:

- **Parametric**: Fixed number of parameters (weights). Learning = adjusting those weights via trial and error.
- **Nonparametric**: Number of parameters grows with the data. Learning = counting patterns.

Deep learning is **parametric**. The number of weights is fixed before training begins.

```python
matrix = [
    ["", "Parametric", "Nonparametric"],
    ["Supervised", "Neural Networks, Linear Regression", "K-Nearest Neighbors, Decision Trees"],
    ["Unsupervised", "K-Means, Autoencoders", "DBSCAN, Hierarchical Clustering"],
]

for row in matrix:
    print(f"{row[0]:<15} {row[1]:<38} {row[2]}")
```

## 4. Supervised Parametric Learning: The Three Steps

Every supervised parametric model follows the same cycle: **Predict → Compare → Learn**.

1. Send input through the model and produce a prediction.
2. Compare the prediction to the true value and measure the error.
3. Adjust the parameters (weights/knobs) to reduce that error.

```python
import numpy as np

np.random.seed(42)
knobs = np.random.rand(3)
inputs = np.array([0.8, 0.3, 0.9])
true_label = 1.0
alpha = 0.1

print("Predict -> Compare -> Learn loop (5 iterations)")
print(f"{'Iter':<6} {'Prediction':>12} {'Error':>12} {'Knobs'}")
print("-" * 60)

for i in range(5):
    prediction = inputs.dot(knobs)
    error = (prediction - true_label) ** 2
    delta = prediction - true_label
    knobs -= alpha * delta * inputs
    print(f"{i+1:<6} {prediction:>12.6f} {error:>12.6f} {np.round(knobs, 4)}")
```

## 5. Nonparametric Learning: Counting Example

A nonparametric model counts patterns directly. The number of parameters equals the number of unique patterns observed.

```python
observations = [
    ("red", "stop"),
    ("green", "go"),
    ("green", "go"),
    ("yellow", "go"),
    ("red", "stop"),
    ("yellow", "stop"),
    ("green", "go"),
]

counts = {}
for color, action in observations:
    if color not in counts:
        counts[color] = {"go": 0, "stop": 0}
    counts[color][action] += 1

print("Nonparametric counting model:")
print(f"{'Light':<10} {'Go':>6} {'Stop':>6} {'P(go)':>8}")
print("-" * 35)
for color, c in counts.items():
    total = c["go"] + c["stop"]
    print(f"{color:<10} {c['go']:>6} {c['stop']:>6} {c['go']/total:>8.2f}")

print(f"\nNumber of parameters: {sum(len(v) for v in counts.values())} (grows with data)")
```
