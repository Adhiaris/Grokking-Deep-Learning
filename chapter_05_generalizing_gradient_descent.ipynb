# Chapter 5: Learning Multiple Weights at a Time
### Generalizing Gradient Descent

Chapter 4 showed gradient descent for a single weight. This chapter generalizes it to any number of weights — multiple inputs, multiple outputs, and both combined.

## 1. Gradient Descent with Multiple Inputs

With multiple inputs there is one `weight_delta` per weight. Each is computed by multiplying the shared output `delta` by the weight's own input value.

```python
import numpy as np

weights = np.array([0.1, 0.2, -0.1])
alpha   = 0.01

toes  = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]
win_or_lose_binary = [1, 1, 0, 1]

input_vec = np.array([toes[0], wlrec[0], nfans[0]])
true      = win_or_lose_binary[0]

print(f"{'Iter':<6} {'Prediction':>12} {'Error':>12} {'Weights'}")
print("-" * 65)

for iteration in range(5):
    pred         = input_vec.dot(weights)
    error        = (pred - true) ** 2
    delta        = pred - true
    weight_deltas = delta * input_vec
    weights      -= alpha * weight_deltas
    print(f"{iteration:<6} {pred:>12.6f} {error:>12.6f} {np.round(weights, 5)}")
```

## 2. Freezing One Weight

If one weight is frozen (its delta is forced to 0), the other weights still converge because all weights share the same error measure. However, the frozen weight loses its chance to contribute to prediction.

```python
import numpy as np

weights = np.array([0.1, 0.2, -0.1])
alpha   = 0.3

input_vec = np.array([8.5, 0.65, 1.2])
true      = 1

print("Training with weight[0] frozen:")
print(f"{'Iter':<6} {'Error':>12} {'Weights'}")
print("-" * 55)

for iteration in range(5):
    pred          = input_vec.dot(weights)
    error         = (pred - true) ** 2
    delta         = pred - true
    weight_deltas = delta * input_vec
    weight_deltas[0] = 0
    weights      -= alpha * weight_deltas
    print(f"{iteration:<6} {error:>12.6f} {np.round(weights, 4)}")
```

## 3. Gradient Descent with Multiple Outputs

With multiple outputs there is one `weight_delta` per output. Each is computed by multiplying the single shared input by each output's own `delta`.

```python
import numpy as np

weights = np.array([0.3, 0.2, 0.9])
alpha   = 0.1

wlrec = [0.65, 1.0, 1.0, 0.9]
hurt  = [0.1, 0.0, 0.0, 0.1]
win   = [1,   1,   0,   1  ]
sad   = [0.1, 0.0, 0.1, 0.2]

input_val = wlrec[0]
true      = np.array([hurt[0], win[0], sad[0]])

print(f"{'Iter':<6} {'Error':>12} {'Weights'}")
print("-" * 55)

for iteration in range(6):
    pred          = input_val * weights
    error         = ((pred - true) ** 2).sum()
    delta         = pred - true
    weight_deltas = input_val * delta
    weights      -= alpha * weight_deltas
    print(f"{iteration:<6} {error:>12.6f} {np.round(weights, 4)}")
```

## 4. Gradient Descent with Multiple Inputs and Outputs

Combining both cases: the weight deltas form a matrix (outer product of input and delta vectors).

```python
import numpy as np

weights = np.array([
    [0.1,  0.1, -0.3],
    [0.1,  0.2,  0.0],
    [0.0,  1.3,  0.1],
])
alpha = 0.01

toes  = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]
hurt  = [0.1, 0.0, 0.0, 0.1]
win   = [1,   1,   0,   1  ]
sad   = [0.1, 0.0, 0.1, 0.2]

input_vec = np.array([toes[0], wlrec[0], nfans[0]])
true      = np.array([hurt[0], win[0], sad[0]])

print(f"{'Iter':<6} {'Total Error':>14}")
print("-" * 25)

for iteration in range(6):
    pred          = input_vec.dot(weights.T)
    error         = ((pred - true) ** 2).sum()
    delta         = pred - true
    weight_deltas = np.outer(input_vec, delta)
    weights      -= alpha * weight_deltas.T
    print(f"{iteration:<6} {error:>14.6f}")
```

## 5. MNIST Preview: Visualizing What Weights Learn

When training on image data (like MNIST's 28×28 handwritten digits), the weights for each output node form a pattern that resembles the shape of the digit they predict. This is because the dot product is a similarity measure — high weights activate most strongly for inputs that look like the pattern.

```python
import numpy as np

np.random.seed(0)

weights = np.random.rand(784, 10) - 0.5
print("Weight matrix shape:", weights.shape)
print("Inputs : 784 (28x28 pixels)")
print("Outputs: 10  (one per digit)")
print()

fake_image = np.random.rand(784)
scores = fake_image.dot(weights)
predicted_digit = scores.argmax()

print("Raw output scores:", np.round(scores, 2))
print("Predicted digit  :", predicted_digit, "(untrained — random)")
```
