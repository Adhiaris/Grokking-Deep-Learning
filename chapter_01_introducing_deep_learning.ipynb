## 6. A Neural Network in One Line

The book promises you will build your first neural network a few pages into Chapter 3. Below is a preview of what that looks like at its most fundamental level — a single input multiplied by a single weight to produce a prediction.

```python
import numpy as np

weight = 0.1

def neural_network(input_val, weight):
    return input_val * weight

number_of_toes = [8.5, 9.5, 10.0, 9.0]

print("Input (toes) -> Prediction (win probability)")
print("-" * 45)
for toes in number_of_toes:
    prediction = neural_network(toes, weight)
    print(f"  {toes} toes  ->  {prediction:.3f}")
```

## 7. Learning: The Core Idea

A neural network learns by adjusting its weights. If a prediction is too high, the weight decreases. If a prediction is too low, the weight increases. This is the essence of the learning loop that the rest of the book builds upon.

The full cycle is: **Predict → Compare → Learn**.

```python
import numpy as np

weight = 0.5
input_val = 0.8
goal = 1.0
learning_rate = 0.01

print(f"{'Step':<6} {'Weight':>10} {'Prediction':>12} {'Error':>10}")
print("-" * 42)

for step in range(10):
    prediction = input_val * weight
    error = (prediction - goal) ** 2
    delta = prediction - goal
    weight -= learning_rate * delta * input_val
    print(f"{step:<6} {weight:>10.6f} {prediction:>12.6f} {error:>10.6f}")
```
